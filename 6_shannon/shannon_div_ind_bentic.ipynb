### Diversity of species of the benthic data set. This data set is a multivariate data set. The Dutch government institute RIKZ carried out a marine benthic sampling program in the Summer of 2002. Data on 75 marine benthic species were collected at 45 sites at nine beaches along the Dutch coastline.
### The data matrix consists of 45 rows (sites) and 88 colums (75 Species and 13 explanatory variables).
### In this project your task is to calculate some diversity indices given below. A diversity index means, that for each site, you will characterise the 75 species with a single value. The diversity indices are defined:
1. Total abundance per site. This is the total number of marine benthic organisms per site.
2. Species richness, defined as the number of different species per site.
3. Shannon index. This takes into account both the presence/absence nature of the data and the actual abundance it is defined by
$$
H_{i} = - \sum_{j=1}^{m}p_{ij} \ln(p_{ij})
$$
with
$$
p_{ij} = \frac{Y_{ij}}{\sum_{j=1}^{m}Y_{ij}} \quad .
$$

We will be using only these. These are written in the data fram species (45 x 75) . One row for each site, for example species.iloc[0,0] gives you the number 4, which means at site one are 4 organisms of species 1 .
