## Random walk in two dimensions
You will draw random numbers and assign it to $x$ and $y$ coordinates.
The random numbers we draw are from the interval $[0,1)$. We define directions in the following way, if the random number is in one of the following intervals: 
- $[0.00,0.25)$: North
- $[0.25,0.50)$: East
- $[0.50,0.75)$: South
- $[0.75,1.00)$: East

```python
import random as rnd
r = rnd.random()
print(r)
```

### In the above code you see how one can create random numbers in the interval $[0,1)$ , one can do this in a loop now and cann them assign to $x$ and $y$ coordinates according to the above listing. We define a diameter $d=1$ meter for example, if you have to move to east then you add $d$ to your $x$  coordinate. To enable plotting the path, use two arrays for storing the coordinate history! Set the initial positions of the $x$ and $y$ coordinates to zero. Plot the path with a red line, mark the starting point with blue o and the end point with a blue * .
